```python
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
import datetime
from datetime import datetime, date, time, timedelta
%matplotlib inline
from matplotlib.ticker import MaxNLocator
import matplotlib.ticker as ticker
```

```python
# read the dataset
data = pd.read_csv("time_series_data.csv")
```

```python
# convert to the date format
data['new_dates'] = pd.to_datetime(data['new_dates'], format='%Y-%m-%d')
date = data['new_dates']
date
```

```python
# extract the columns to visualize
mean_cases = data['new_cases_mean_data']
upper_cases = data['new_cases_upper_data']
lower_cases = data['new_cases_lower_data']
```

```python
# set the size of the graph
fig_1, ax = plt.subplots(1,1, figsize = (20, 12))

# plot and format the graph
ax.plot(date, upper_cases, '--c', date, lower_cases, '--c',
         date, mean_cases, 'k',)
ax.set_xlabel('Date', fontsize = 20)
ax.set_ylabel('Number of Cases', fontsize = 20)
ax.set_title('Forecast Number of Confirmed Cases in NY', fontsize = 25)

tick_spacing = 5
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

for tick in ax.get_xticklabels():
    tick.set_rotation(40)

plt.setp(ax.get_xticklabels(), fontsize=15)
plt.setp(ax.get_yticklabels(), fontsize=15)

plt.fill_between(date, upper_cases, lower_cases, alpha=0.2)

plt.savefig('time_series_forecast_plot.png', dpi=300)
plt.show()
```
